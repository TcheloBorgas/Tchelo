```python
#━━━━━━━━━❮Bibliotecas❯━━━━━━━━━
import requests
import pandas as pd
import json
#━━━━━━━━━❮◆❯━━━━━━━━━

#━━━━━━━━━❮acessando a api❯━━━━━━━━━
card_aleatorio = requests.get('https://api.scryfall.com/cards/random')
dados =card_aleatorio.json()
#━━━━━━━━━❮◆❯━━━━━━━━━

vlr = {"name" : ["0"],
        "oracle_id" : ["1"],
        "oracle_text":["2"],
        "mana_cost": ["3"],
        "prices_usd" :["4"]}
    
vlr["name"][0] = dados["name"]
vlr["oracle_id"][0] = dados["oracle_id"]
vlr["oracle_text"][0] = dados["oracle_text"]

vlr["mana_cost"][0] = dados["mana_cost"]
vlr["prices_usd"][0] = dados["prices"]["usd"]

bd =[]

for i in range(200):
    vlr = {"name" : ["0"],
        "oracle_id" : ["1"],
        "oracle_text":["2"],
        "mana_cost": ["3"],
        "prices_usd" :["4"]}
    
    try:     
        r = requests.get("https://api.scryfall.com/cards/random")
        data = r.json()
        vlr["name"][0] = data["name"]
        vlr["oracle_id"][0] = data["oracle_id"]
        vlr["oracle_text"][0] = data["oracle_text"]
        vlr["mana_cost"][0] = data["mana_cost"]
        vlr["prices_usd"][0] = data["prices"]["usd"]
        bd.append(vlr) 
    
    except:
        print('Erro')

        
data_brt = pd.DataFrame(bd)
display(data_brt)
    

#━━━━━━━━━❮Gerando arquivo❯━━━━━━━━━
data_brt.to_csv("Magic.csv")
    
    
    
```
