**Questão 3** - O primeiro modelo que vimos, foi o modelo de **regressão linerar**. Imagine que estamos dando uma consultoria para uma empresa que precisa prever a renda dos seus clientes. Use a regressão linerar para realizar essa previsão. Mas antes divida os dados em treino e teste. Sendo **80%** para treino e **20%** para teste

```python
from sklearn.linear_model import LinearRegression
```

```python
treino = regressao.sample(int(len(regressao)*0.8))

X_treino = treino[['idade',	'carros',	'renda',	'Pinheiros',	'Saude']]
y_treino = treino[["renda"]]

teste = regressao[~regressao.index.isin(treino.index)]

X_teste = teste[['idade',	'carros',	'renda',	'Pinheiros',	'Saude']]
y_teste = teste[["renda"]]

```

```python
reg = LinearRegression().fit(X_treino, y_treino)
```

**Questão 4** - Pesquise na documentação o método score e aplique ele para calculcar o R2 na base de testes


https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
treino_score = reg.score(X_treino, y_treino)
teste_score = reg.score(X_teste, y_teste)
print(f"R2 treino {treino_score}, R2 teste {teste_score}")
```
