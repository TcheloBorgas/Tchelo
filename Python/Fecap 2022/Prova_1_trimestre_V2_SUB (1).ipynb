**Questão 6 - (1,5 ponto)**

Agora que você já entendeu como acessar as variáveis dessa API, podemos construir o banco de dados! Para isso você deve substituir ########## pelo código necessário para que o data frama, chamado df seja impresso na tela. **NÃO MECHA EM NENHUMA OUTRA PARTE DO CÓDIGO!**

```python
import pandas as pd

lista = []
i = 1 

while i <= 10:
  resp = requests.get("https://opentdb.com/api.php?amount=10&category=27&type=multiple")
  data = resp.json()

  for q in data['results']:
    values = {"category" : "0",
          "correct_answer" :"1",
          "difficulty": "2",
          "incorrect_answers" :"3",
          "question" : "4",
          "type" : "5"        
        }
    values["category"] = q["category"]
    values["correct_answer"] = q['correct_answer']
    values["difficulty"] = q["difficulty"]
    values["incorrect_answer"] = q["incorrect_answers"]
    values["question"] = q['question']
    values["type"] = q['type']
    lista.append(values) 
 
  i += 1

df = pd.DataFrame(lista)
display(df)
```
